# **1. Importing Libraries**

```python
from PIL import Image
import cv2
from IPython.display import display
import matplotlib.pyplot as plt
```

# **2. Reading the image**

```python
image = cv2.imread("my_img.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

# **3. Greyscale Image**

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Create a PIL image from the grayscale array
pil_image = Image.fromarray(gray_image)

# Display the grayscale image
display(pil_image)
```

#  **4. Inverted Image**

```python

# Invert the grayscale image
inverted_image = 255 - gray_image

# Create a PIL image from the inverted array
pil_image = Image.fromarray(inverted_image)

# Display the inverted image
display(pil_image)
```

# **5. Blurred Image**

```python
# Apply Gaussian blur to the inverted image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)

# Create a PIL image from the blurred array
blurred_image = Image.fromarray(blurred)

# Display the blurred image
display(blurred_image)
```

# **6.Pencil Sketched Image**

```python
# Create the pencil sketch by dividing the grayscale image by the inverted blurred image
pencil_sketch = cv2.divide(gray_image, blurred, scale=256.0)

# Create a PIL image from the pencil sketch array
pil_image = Image.fromarray(pencil_sketch)

# Display the pencil sketch image
display(pil_image)
```

# **7. Displaying Final Results**

```python
# Set up the figure and axes
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle("Image to Pencil Sketch Transformation", fontsize=16)

# Display the original image with label
axes[0, 0].imshow(image)
axes[0, 0].set_title("Original Image")

# Display the grayscale image with label
axes[0, 1].imshow(gray_image, cmap='gray')
axes[0, 1].set_title("Grayscale Image")

# Display the inverted image with label
axes[1, 0].imshow(inverted_image, cmap='gray')
axes[1, 0].set_title("Inverted Image")

# Display the pencil sketch with label
axes[1, 1].imshow(pencil_sketch, cmap='gray')
axes[1, 1].set_title("Pencil Sketch")

# Remove the axis labels
for ax in axes.flatten():
    ax.axis('off')

# Adjust spacing and layout
plt.tight_layout()

# Show the plot
plt.show()
```
